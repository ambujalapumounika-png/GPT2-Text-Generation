```python
import markovify
```

```python
text = """
Artificial Intelligence is changing technology.
AI helps doctors diagnose diseases.
Machine learning improves prediction systems.
Deep learning is used in image recognition.
AI assists businesses in decision making.
AI helps people solve complex problems.
Neural networks learn patterns from data.
Data science combines statistics and computing.
AI is transforming industries around the world.
Technology is growing rapidly with AI.
Machine learning supports data analysis.
AI improves productivity and innovation.
"""
```

```python
text_model = markovify.Text(text, state_size=1)
```

```python
for i in range(5):
      print(text_model.make_short_sentence(100))
```
